```python
from shapely.geometry import LineString, Polygon, mapping, asShape
from scipy.interpolate import splprep, splev
import fiona
import numpy as np
```

```python
src_file = "data/vl_riks.shp"
dst_file = "data/spline_vl_riks.shp"
```

```python
#Read File and file properties, iterate and select first polyline
with fiona.open(src_file) as src_fl:
    src_vals = src_fl.values()
    val1 = next(src_vals)
    val2 = next(src_vals)
    val3 = next(src_vals)
    geom = asShape(val1['geometry'])    
```

```python
geom
```

```python
#Get Coordinates
if isinstance(geom, LineString):
    x = np.array(geom.coords.xy[0])
    y = np.array(geom.coords.xy[1])

elif isinstance(geom, Polygon):
    x = np.array(geom.exterior.coords.xy[0])
    y = np.array(geom.exterior.coords.xy[1])
    
coords = np.array([x,y])
```

```python
#spline prep
'''
t - ndarray, shape (n+k+1,)
knots

c - ndarray, shape (>=n, â€¦)
spline coefficients

k - int
B-spline degree
'''
tck_u, fp, ier, msg = splprep([x,y], s = 0, k = 2, nest=-1, full_output=1)

n_len = len(x)*3

x_ip, y_ip = splev(np.linspace(0, 1, n_len),tck_u[0])
coords_ip = np.array([x_ip, y_ip])
```

```python
#Generate LineString/Polygon from coords
if isinstance(geom, LineString):
            geom_ip = LineString(coords_ip.T)

elif isinstance(geom, Polygon):
            geom_ip = Polygon(coords_ip.T)
```

```python
geom_ip
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fig.set_size_inches(12, 6)
plt.title('Interpolation Result (x3)')
ax.plot(x,y, 'b',lw=2, label='Polyline')
ax.plot(x,y, 'ob',lw=2, label='Polyline Vertex')
ax.plot(x_ip,y_ip, 'r', lw=2, alpha=0.7, label='Interpolated Spline')
ax.plot(x_ip,y_ip, 'or', alpha = 0.2, label='Spline Points')
#ax.plot(tck_u[1],'go', label = 'knots')
ax.grid(True)
ax.legend(loc='best')
plt.show()
```
